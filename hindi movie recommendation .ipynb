```python
data_features=data[['Genre','Cast_stars','Crew_dir']].fillna('')
```

```python
x=data_features['Genre']+' '+data_features['Cast_stars']+' '+data_features['Crew_dir']
```

```python
x.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf=TfidfVectorizer()
```

```python
x=tf.fit_transform(x)
```

```python
print(x)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cs=cosine_similarity(x)
```

```python
cs
```

```python
cs.shape
```

```python
movie_name=input("enter movie name: ")
```

```python
movie_title=data['Title'].tolist()
```

```python
movie_title
```

```python
import difflib as dl
```

```python
recommendation = dl.get_close_matches(movie_name,movie_title)
print(recommendation)
```

```python
close_match=recommendation[0]
print(close_match)
```

```python
index_number=data[data.Title == close_match]['Movie_id'].values[0]
print(index_number)
```

```python
recommendation_score=list(enumerate(cs[index_number]))
print(recommendation_score)
```

```python
len(recommendation_score)
```

```python
sorted_similar=sorted(recommendation_score,key= lambda x:x[1],reverse=True)
print(sorted_similar)
```

```python
print("these are the top 30 movies recommend for you\n")
i=1
for movie in sorted_similar:
    index=movie[0]
    title_from_idx=data[data.index==index]['Title'].values[0]
    if (i<11):
        print(i," ",title_from_idx)
        i+=1
```
